### Part of speech

There is no feature for the part of speech of a word, but you can find them using a regular expression with the feature morph. The start signs indicate the part of speech of a word:

- A- adjective
- C- conjunction
- D- adverb
- I- interjection
- N- noun
- P- preposition
- RA definite article
- RD demonstrative pronoun
- RI interrogative/indefinite pronoun
- RP personal pronoun
- RR relative pronoun
- V- verb
- X- particle
